**Dropping un required columns**

```python
dtc_data= dtc_data.drop('Id', axis=1)
```

```python
dtc_data
```

**Data Preprocessing**

```python
from sklearn.preprocessing import LabelEncoder    #used to standardize the categorial value
le=LabelEncoder()
```

```python
dtc_data['Species']=le.fit_transform(dtc_data['Species'])   #changing categorial values into numbers
```

```python
dtc_data
```

**Spliting the data**

```python
x=dtc_data.iloc[:,:-1]
y=dtc_data.iloc[:,-1]
x.head()
```

```python
y.head()
```

****Train and Test split****

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test= train_test_split(x,y,test_size=0.20,random_state=50)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
x_train.shape
```

```python
y_test.shape
```

**DECISION TREE ALGORITHM**

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
y_predict=dtc.predict(x_test)
```

```python
y_test=np.array(y_test)
y_test
```

**Accuracy**

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix


#accuracy and confusion matrix shows how good is our prediction
confusion_matrix(y_predict,y_test)         #more values in diagonal elements more the accuracy
acc=accuracy_score(y_predict,y_test)*100    #near to 1 / 100% is considered as best prediction
acc
```

# **Classificion Report**

```python
from sklearn.metrics import classification_report
print(classification_report(y_predict,y_test))
```

***DESICION TREE VISUALIZATION***

```python
from sklearn import tree
```

```python
plt.figure(figsize=(20,20))
dtv= tree.plot_tree(dtc, feature_names =x.columns,filled=True,fontsize=15)
```
